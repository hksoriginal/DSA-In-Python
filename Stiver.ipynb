# Pascal Triangle


```python
class solution:
    def generate(self,numRows):
        res = [[1]]
        for i in range(numRows-1):
            temp = [0] + res[-1] + [0]
            rows = []
            for j in range(len(res[-1])+1):
                rows.append(temp[j]+temp[j+1])
            res.append(rows)
        return res
            
```

```python
s= solution()
```

```python
pas = s.generate(6)
pas
```

```python
for i in pas:
    for j in i:
        print(j,end =' ')
    print()
```
