```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def insert(self,newNode):
        if self.head is None:
            self.head = newNode
        else:
            x = self.head
            while True:
                if x.next is None:
                    break
                x=x.next 
            x.next = newNode
            
    def display(self):
        current = self.head
        while True:
            if current is None:
                break
            print(current.data,end=" ")
            current = current.next
    def count(self):
        current = self.head
        self.count_node = 0
        while True:
            if current is None:
                break
            current = current.next
            self.count_node +=1
        print('\nTotal Nodes : ',self.count_node,end = " ")
        
    def insert_at_end(self,newNode):
        self.insert(newNode)
        
    def insert_at_begin(self,newNode):
        current = self.head
        newNode.next = current
        self.head = newNode
        
    def insert_at_a_pos(self,newNode,n):
        current = self.head
        index = 1
        while True:
            if current is None:
                break
            index+=1
            if index == n:
                newNode.next=current.next
                current.next = newNode
            current = current.next
            
    def deletion_at_begining(self):
        self.head = self.head.next
        
    def deletion_at_end(self):
        current = self.head
        while current.next.next is not None:
            current = current.next
        current.next = None
    
    def deletion_at_position(self,n):
        current = self.head
        index = 1
        while True:
            index+=1 
            if index == n:
                current.next = current.next.next
                break
            current = current.next
    def reverse(self):
        if self.head.next == None:
            return
        p1 = self.head
        p2 = p1.next
        p3 = p2.next
        p2.next = p1
        p1.next = None
        self.head = p2
            
        
    
    
            
        
                
    
                
                
            

```

```python
# node in a linkedList
def getNode(head,positionFromTail):
    ptr = head
    count = 0
    while ptr is not None:
        count+=1
        ptr = ptr.next
    ptr = head
    count-=1
    while ptr.next is not None:
        if count==positionFromTail:
            break
        ptr = ptr.next
        count-=1
    return ptr.data
```

```python
# 2nd approach
def getNode(head,positionFromTail):
    ptr1 = head
    ptr2 = head
    count = 0
    while ptr1.next is not None:
        ptr1 = ptr1.next
        count+=1
        if count > positionFromTail:
            ptr2=ptr2.next
    return ptr2.data
```

```python
#compare two linkedlist

def compare_lists(llist1, llist2):
    p1 = llist1
    p2 = llist2
    flag = True
    while p1 != None and p2!=None:
        if p1.data == p2.data:
            p1 = p1.next
            p2 = p2.next
            continue
        flag = False
        break
    if(p1!=None and p2==None) or (p1==None and p2!=None):
         flag = False
    if flag:
        return 1
    return 0
        
            
```

```python
# delete duplicate nodes
def removeDuplicates(llist):
    ptr = llist
    while ptr.next == ptr.next.data:
        ptr.next = ptr.next.next
        continue
        ptr = ptr.next
    return llist
        
```

```python
# find merge point of two lists
def findMergeNode(head1, head2):
    p1 = head1
    p2 = head2
    while 1:
        if p1 == p2:
            break
        p1 = p1.next
        p2 = p2.next
        if p1==None:
            p1 = head2
        if p2 ==None:
            p2 = head1
    return p1.data 
        
```

```python
#reverse a Linked List using pointer
def reverse(head):
    p1 = head
    p2 = p1.next
    p3 = p2.next
    p2.next = p1
    p1.next = None
    while p3 is not None:
        p1 = p2
        p2 = p3
        p3 = p3.next
    return p2   
        
```

```python
#reverse a Linked List using recursion
def reverse(head):
    if head.next == None:
        return head

    p1 = reverse(head.next)
    head.next.next = head
    head.next=None
    return p1
    
```

```python
#cycle detection in linkedList
def has_cycle(head):
    visited = []
    while  head is not None:
        if head in visited:
            return True
        else:
            visited.append(head)
            head = head.next
    return False
    
```
