```python
# balanced Brackets
def check(s):
    lst = list()
    for i in range(len(s)):
        if s[i] in ('({[]})'):
            lst.append(s[i])
        if (len(lst) > 1 and ((lst[-2]=='[' and lst[-1]==']') or (lst[-2]=='(' and lst[-1]==')')or(lst[-2]=='{' and lst[-1]=='}'))):
            lst.pop()
            lst.pop()
    if len(lst) == 0:
        print('YES')
    else:
        print('NO')
            
        
            
            
        
for _ in range(int(input())):
    check(input())
```

```python
#waiter Problem
prime_num = []

def isprime(n):
    for j in range(2,int(n/2)+1):
        if n%j==0:
            return False
        return True

for n in range(2,10000):
    if (isprime(n)):
        prime_num.append(n)
        
        
def waiter(number,q):
    A = []
    B = []
    answers = []
    for i in range(q):
        while(number):
            value = number.pop()
            if value % prime_num[i] == 0:
                B.append(value)
            else:
                A.append(value)
        while(B):
            answers.append(B.pop())
        number = A
        A = []
    while(number): 
        answers.append(number.pop())
    return answers
    
        
    
```
